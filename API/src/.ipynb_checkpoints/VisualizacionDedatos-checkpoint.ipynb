```python
#quitar el espacio y volverlo a meter en el dataset
def quitaespacios(campo):
    valor_trip = df[campo].str.strip()
    df[campo] = valor_trip
quitaespacios("stage")
quitaespacios("team_name_home")
quitaespacios("team_name_away")

df.to_csv("EurocopaLimpio.csv")
```
